```python
def generate_resume(name, role, skills, education, projects):
    resume = f"""
NAME: {name}

ROLE:
{role}

SKILLS:
{skills}

EDUCATION:
{education}

PROJECTS:
{projects}

SUMMARY:
{name} is a motivated individual aiming for a career as a {role}.
Skilled in {skills}, with academic background in {education}.
"""
    return resume
```

```python
print(generate_resume(
    "hafsal",
    "iot Developer",
    "Python, AI, ML",
    "BCA",
    "AI Resume Builder Project"
))
```
